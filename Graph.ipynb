```python
x1=[1,3,5,7]
y1=[5,2,6,6]
x2=[2,4,6,8]
y2=[6,3,5,2]
plt.bar(x1,y1,label='first')
plt.bar(x2,y2,label='second')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('bar chart')
plt.show()
```

```python
sale_value=[12,10,5,7,0,14,15,24,10,11,15,30,18,10,11,18,30]
bins=[0,5,10,15,20,25,30]
plt.hist(sale_value,bins,histtype='bar',rwidth=0.6)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
def sinplot():
    x = np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*.5)*(7-i))
sinplot() 
plt.show()
    
```
