```python
x=data['Head Size(cm^3)'].values
y=data['Brain Weight(grams)'].values
```

```python
# Mean of x and y
mean_x=np.mean(x)
mean_y=np.mean(y)

#Total number of values
m=len(x)

#Using the formula to calculate b1 and b2
number=0
denom=0
for i in range(m):
    number+=(x[i]-mean_x)*(y[i]-mean_y)
    denom+=(x[i]-mean_x)**2
b1=number/denom
b0=mean_y-(b1*mean_x)

#Print coefficients
print(b1,b0)
```

```python
y_pred =b0+b1*x
```

```python
rmse=0
for i in range(m):
    y_pred=b0+b1*x[i]
    rmse+=(y[i]-y_pred)**2
rmse=np.sqrt(rmse/m)
print(rmse)
```

```python
ss_t=0
ss_r=0
for i in range(m):
    y_pred=b0+b1*x[i]
    ss_t+=(y[i]-mean_y)**2
    ss_r+=(y[i]-y_pred)**2
r2=1-(ss_r/ss_t)
print(r2)
```
